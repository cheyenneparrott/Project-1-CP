```python
#Import modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import json
import os
```

```python
#Read by count
us_by_count = pd.read_csv("US_CRE_by_count.csv")
us_by_count.head(2)
```

```python
#Filter Kansas and New Jersey
ks_nj_by_count = pd.DataFrame(us_by_count.loc[(us_by_count["stname"] == "Kansas") | (us_by_count["stname"] == "New Jersey"), :])
```

```python
#Write to csv file
ks_nj_by_count.to_csv('ks_nj_CRE_by_count.csv')
```
